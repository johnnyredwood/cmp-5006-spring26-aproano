# The Vigenère Cipher

To use the Vigenère Square (or Tabula Recta) to encrypt a message, you start with a big table—26 rows and 26 columns—where each row is the alphabet shifted over by one letter from the previous row. 

First, write out your plaintext. Above it, repeat your keyword until it lines up with every letter in your message. Then, for each letter, find the row for the plaintext letter and the column for the keyword letter. Where they cross, that’s your encrypted letter. Do this for every letter, and you’ve got your ciphertext.

The Vigenère cipher’s big trick is that it doesn’t use just one substitution alphabet like the Caesar cipher. It uses a more than one, switching between them based on your keyword. That means the same plaintext letter can turn into different ciphertext letters, depending on its spot in the message. So, if someone tries to crack the code by looking at letter frequencies, they won't be able to . The patterns get scrambled, and letter frequencies disappear. As long as you use a long, non-repeating keyword, this cipher is much harder to break than a basic substitution cipher.

The Kasiski examination is the classic way people attack the Vigenère cipher. You need to look for repeated sequences of letters in the ciphertext. When you spot them, measure how far apart they are. Then, figure out the greatest common divisors (GCDs) of those distances. The most common GCDs probably match the keyword length. Once you know that, you can split the ciphertext into sections and treat each like its own simple substitution cipher. At that point, common frequency analysis can do the rest.

# The Hill Cipher

### Mathematical Basis: This cipher is based on linear algebra. Explain how a plaintext block is converted into a vector and multiplied by a key matrix.

The Hill cipher is a polyalphabetic substitution cipher that relies on concepts from linear algebra, specifically matrix multiplication in modular arithmetic.

The encryption process works as follows:

Assign each letter of the alphabet a numerical value (ej. A = 0, B = 1, …, Z = 25).

Choose a block size 
𝑛, which determines the dimensions of the key matrix.

Divide the plaintext into blocks of 𝑛 letters.

Convert each plaintext block into an 
𝑛×1
column vector of numbers.

Multiply this plaintext vector by the 
𝑛×𝑛
key matrix.

Apply modulo 26 to each resulting entry to map values back into the alphabet.

Mathematically:

where:

Each ciphertext block is then converted back into letters to form the encrypted message.

# The Playfair Cipher

### Mechanism: Explain the process of creating the 5×5 grid and the specific rules for encrypting a digraph (pair of letters).

To create the 5x5 grid, you have to choose a keyword with no repeating letters. You put that keword, left to right starting from the first row, and if it overflows, you continue on to the next row. Then you fill the spaces in the same manner, left to right, with the remaining letters of the alphabet, and you merge the letters I and J into one for this process.

Then, you "prepare" your text. Whatever message you want to encrypt, you separate into pairs and eliminate spaces. If there are two letters that are the same next to each other, then you put the letter X in between them. For examples:

Original text: Hello
Prepared text: He lx lo

And, just in case there are an uneven number of letters in your message, you fill the space of the last pair with an X.

Afterwards, you go to the grid, pair by pair, putting the letters of each pair in it and follow one of this three rules:

- 1. If the two letters of the pair are placed in the same column, you convert each of them to the letter right below them. And if the letter is at the bottom of the grid, then the process wraps around to the top, and the letter at the top converts to the letter at the top of the column
 - 2. If the two letters of the pair are places in the same row, you convert each of them to the letter at the right of them. And if one letter is in the far right side of the grid, then the process wraps around and that letter becomes the letter at the left of the row.  
 - 3. And finally, if the pair of letters placed in the grid don'w fall in the same row or column, you make a rectangle around them, with them being the diagonally opposite corners of the rectangle, and then you convert each letter, to the corresponding opposite corner of the rectangle following the same row they are in.  

### Historical Context: Research its use by British forces in WWI and WWII. Why was it considered "field-ready" compared to more complex systems?
The Playfair cipher was adopted by British military forces during World War I and continued limited use into World War II, particularly by the British Army and Royal Navy.

It was considered “field-ready” for several practical reasons:

Manual efficiency: It could be encrypted and decrypted quickly with pencil and paper, without mechanical devices.

Stronger than monoalphabetic ciphers: By encrypting letter pairs rather than single letters, it significantly reduced the effectiveness of simple frequency analysis.

Low training overhead: Soldiers could be trained to use it reliably in a short time.

Operational robustness: It functioned in adverse conditions (combat, low light, limited time) where more complex systems were impractical.

While not cryptographically strong by modern standards, Playfair struck a balance between security and usability, making it suitable for tactical communications in the field.

# The Enigma Machine

### Rotor Logic: Explain the concept of the rotors and how they caused the substitution alphabet to change with every single keystroke.
The rotors in the enigma machine had one position for each character of the alphabet, and each key stroke of the machine, made at least one of the rotors change position. Meaning that with each stroke of a letter, the internal mapping of the machine of one letter to another, changed. 
The rotors also had a mechanism that, when one of rotos passes a specific position, it would move the position of the rotor right next to it, again, changing the mapping of the substitution.
